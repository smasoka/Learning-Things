Let get some data

```python
import pandas as pd
df = pd.read_csv("http://bit.ly/tcsv19")
```

Remember the fig and axes exist on the cell it was created on

```python
fig, ax = plt.subplots()
_ = ax.set_title("Titanic Data Set")
_ = ax.plot(np.sort(df['age']), marker='o', markersize=1)
_ = ax.set_ylabel("Ages")
_ = ax.set_xlabel("Passanger")
```

Introduce some bining on the y axes. Set the ticks!

```python
fig, ax = plt.subplots()
_ = ax.set_title("Titanic Data Set")
_ = ax.plot(np.sort(df['age']), marker='o', markersize=1)
_ = ax.set_ylabel("Ages")
_ = ax.set_yticks([15, 25, 55, 64])
_ = ax.set_xlabel("Passanger")
```

```python
fig, ax = plt.subplots()
_ = ax.set_title("Titanic Data Set")
_ = ax.plot(np.sort(df['age']), marker='o', markersize=1)
_ = ax.set_ylabel("Ages")
_ = ax.set_yticks([15, 25, 55, 64])
_ = ax.set_yticklabels(['children', 'youth', 'adults', 'seniors'])
```

Lets do an Exercise

```python
fig, ax = plt.subplots()
_ = ax.set_title("Titanic")
_ = ax.plot(np.sort(df['fare']), marker='o', markersize=1)
_ = ax.set_ylabel("Fare")
_ = ax.set_xlabel("People")
_ = ax.set_yticks([3,12,30,870])
_ = ax.set_yticklabels(["3rd", "2nd", "1stb", "1sts"])
```

Okay, Let's put more than one image on the page

```python
fig, (ax1, ax2) = plt.subplots(ncols=2)
_ = fig.suptitle('Next to (1x2)')
```

Lets put graphs in rows as well

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=2, constrained_layout=True)
_ = fig.suptitle('Next to (1x2)')
```

Okay so how do we plot on the multiple axes ????

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, constrained_layout=True)
_ = fig.suptitle('Titanic Data Set')
_ = ax1.set_title("Fig 1")
_ = ax1.set_ylabel('Y1 Label')
_ = ax1.set_xlabel('X1 Label')

_ = ax2.set_title("Fig 2")
_ = ax2.set_ylabel('Y2 Label')
_ = ax2.set_xlabel('X2 Label')
```
