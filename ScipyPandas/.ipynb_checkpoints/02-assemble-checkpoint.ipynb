#### Adding new row

```python
new_row = pd.Series(['n1', 'n2', 'n3', 'n4'])
```

```python
new_row
```

```python
pd.concat([df1, new_row])
```

```python
# Double square brackets
new_row_2 = pd.DataFrame([['n1', 'n2', 'n3', 'n4']], columns=['A', 'B', 'C', 'D'])
```

```python
new_row_2
```

```python
pd.concat([df1, new_row_2])
```

Unequal rows or dataframe

```python
new_row_3 = pd.DataFrame([['n1', 'n2', 'n3']], columns=['A', 'B', 'C'])
new_row_3
```

```python
pd.concat([df1, new_row_3], sort=False)
```

```python
new_row_4 = pd.DataFrame([['n1', 'n2', 'n3', 'n4', 'n5', 'n6']], \
                         columns=['A', 'B', 'C', 'D', 'E', 'F'])
new_row_4
```

```python
pd.concat([df1, new_row_4], sort=False)
```

```python
# Renaming Columns
df1.columns = ['A', 'B', 'C', 'D']
df2.columns = ['A', 'E', 'F', 'G']
```

```python
# Realignment, rows sort themselves
pd.concat([df1, df2], axis=1, sort=False)
```

### Merging Data

```python
person = pd.read_csv('data/survey_person.csv')
site = pd.read_csv('data/survey_site.csv')
survey = pd.read_csv('data/survey_survey.csv')
visited = pd.read_csv('data/survey_visited.csv')
```

```python
person
```

```python
site
```

```python
survey.head()
```

```python
visited
```

```python
visited_subset = visited.iloc[[0, 2, 6]]
visited_subset
```

```python
# "site" uses 'name' and "visited_subset" uses 'site' as keys
# by default, it does "inner join"
o2o = pd.merge(site, visited_subset, left_on='name', right_on='site')
o2o
```

```python
# Many to One
# Notice the missing DataSet
m = site.merge(visited, left_on='name', right_on='site')
m
```

```python
df1 = pd.DataFrame({
    'a': [1,1,1,2,2],
    'b': [10,20,30,40,50]
})

df2 = pd.DataFrame({
    'a1': [1,1,2,2,3],
    'b1': [100,200,300,400,500]
})
```

```python
df1
```
