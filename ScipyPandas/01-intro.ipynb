#### Groupings

```python
# Returns a DataFrame GroupBy 
df.groupby('year')
```

```python
# Returns a Series Group By (1 dimentional)
df.groupby('year')['lifeExp']
```

```python
# df['lifeExp'].mean() -- giving us the mean for entire dataset
df.groupby('year')['lifeExp'].mean()
```

```python
# Group by Year
# Calc lifeExp and gdpPercap
df.groupby('year')[['lifeExp', 'gdpPercap']].mean()
```

```python
# Group by Year first
# Group by continent next
# Calc lifeExp and gdpPercap
df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean().head()
```

```python
# nunique 
df.groupby('continent')['country'].nunique()
```

```python
gyle = df.groupby('year')['lifeExp'].mean()
```
